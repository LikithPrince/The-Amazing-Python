```python
blank=np.zeros((500,500,3),dtype='uint8')
# blank.shape
flag=True
counter=0
while flag:
    
    cv2.line(blank,(0+counter,0),(blank.shape[0]-counter,blank.shape[1]),(0,0,255),thickness=2)
    cv2.line(blank,(0,0+counter),(blank.shape[0],blank.shape[1]-counter),(255,0,0),thickness=2)
    counter+=50
    if counter>500:
        flag=False
cv2.imshow('Blank',blank)
cv2.waitKey()
```

Reading an Image

```python
import cv2
image=cv2.imread('Sales Demo Front.png')
h , w =image.shape[:2]
print(f"Height= {h}, Width= {w}")
```

Extracting the RGB value of a pixel

Note- OpenCV arranges the channels in BGR order

```python
(B,G,R)=image[100,100]
print(f"B={B},G={G},R={R}")

# WE CAN GET INDIVIDUAL VALUE ALSO
B=image[100,100,0]
B
```

Extracting the Region of Interest (ROI)

```python
roi = image[0 : w, 0 : h]
roi
```

Resizing the Image

```python
resize_image=cv2.resize(image,(1000,1000))
cv2.imwrite("resized image.png",resize_image)
resize_image.shape
```

To maintain the aspect ratio

```python
ratio=1000/w
dim=(1000,int(h*ratio))
resize_aspect=cv2.resize(resize_image,dim)
resize_aspect.shape
# cv2.imwrite("aspect ratio image.png",resize_aspect)
```

Rotating the Image



```python
center=(w//2,h//2)
matrix=cv2.getRotationMatrix2D(center,45,1)
rotated=cv2.warpAffine(image,matrix,(w,h))
cv2.imwrite("rotated image.png",rotated)
```

Drawing a Rectangle

```python
output=image.copy()
rect=cv2.rectangle(output,(950,1280),(2270,2240),(0,0,255),10)
cv2.namedWindow('Rectangle Image', cv2.WINDOW_NORMAL)
cv2.imshow("Rectangle Image",rect)
cv2.waitKey()
```

Displaying text

```python
output_image=image.copy()
text=cv2.putText(output_image,'Demokit',(1270,1200),cv2.FONT_HERSHEY_COMPLEX,5,(0,0,255),20)
cv2.namedWindow('Rectangle Image', cv2.WINDOW_NORMAL)
cv2.imshow("Rectangle Image",text)
cv2.waitKey()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# img=cv2.imread("Sales Demo Front.png")
img = mpimg.imread("aspect ratio image.png")
# print(img)
kernal=np.ones((5,5), np.uint8)
erosion=cv2.erode(img,kernal,iterations=1)
imgplot = plt.imshow(img)
```

```python
imgplot = plt.imshow(erosion)
```

```python
dilation=cv2.dilate(img,kernal,iterations=1)
imgplot = plt.imshow(dilation)
```

```python
opening=cv2.morphologyEx(img,cv2.MORPH_CROSS,kernal)
imgplot=plt.imshow(opening)
```

```python
closing=cv2.morphologyEx(img,cv2.MORPH_CROSS,kernal)
imgplot=plt.imshow(closing)
```

```python
gradient=cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernal)
imgplot=plt.imshow(gradient)
```

```python
tophat=cv2.morphologyEx(img,cv2.MORPH_TOPHAT,kernal)
imgplot=plt.imshow(tophat)
```

```python
blackhat=cv2.morphologyEx(img,cv2.MORPH_BLACKHAT,kernal)
imgplot=plt.imshow(blackhat)
```

```python
# Rectangular Kernel
rect_kar=cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
```

```python
# Elliptical Kernel
elli_kar=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(15,7))
```

```python
# Cross-shaped Kernel
cross_kar=cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5))
```

```python
closing=cv2.morphologyEx(img,cv2.MORPH_CLOSE,cross_kar)
# imgplot=plt.imshow(closing)
plt.subplot(1,1,1),plt.imshow(closing,vmin=0,vmax=255)
plt.xticks([]),plt.yticks([])
```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread('Sales_Demo_Front.png',0)
ret,thresh1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
ret,thresh3 = cv.threshold(img,127,255,cv.THRESH_TRUNC)
ret,thresh4 = cv.threshold(img,127,255,cv.THRESH_TOZERO)
ret,thresh5 = cv.threshold(img,127,255,cv.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
thresh1.shape
```
